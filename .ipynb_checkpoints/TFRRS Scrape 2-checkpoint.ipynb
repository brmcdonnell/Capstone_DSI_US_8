```python
# Preparing for Purdy points function
def frac(d):
    if (d <110):
        return 0
    else:
        laps = int(d/400)
        meters = d - laps*400
        if (meters <= 50):
            partlap = 0
        elif (meters <= 150):
            partlap = meters - 50
        elif (meters <=250):
            partlap = 100
        elif (meters <=350):
            partlap = 100 + (meters - 250)
        elif (meters <=400):
            partlap = 200
    tmeters= laps*200 + partlap
    return (tmeters/d)
```

```python
# Purdy points function
def purdy(dist, tsec):
    i = 0
    ptable= [40.0,11.000, 50.0,10.9960, 60.0,10.9830, 70.0,10.9620,
            80.0,10.934, 90.0,10.9000,100.0,10.8600,110.0,10.8150,
            120.0,10.765,130.0,10.7110,140.0,10.6540,150.0,10.5940,
            160.0,10.531,170.0,10.4650,180.0,10.3960,200.0,10.2500,
            220.0,10.096,240.0, 9.9350,260.0, 9.7710,280.0, 9.6100,
            300.0, 9.455,320.0, 9.3070,340.0, 9.1660,360.0, 9.0320,
            380.0, 8.905,400.0, 8.7850,450.0, 8.5130,500.0, 8.2790,
            550.0, 8.083,600.0, 7.9210,700.0, 7.6690,800.0, 7.4960,
            900.0,7.32000, 1000.0,7.18933, 1200.0,6.98066, 1500.0,6.75319,
            2000.0,6.50015, 2500.0,6.33424, 3000.0,6.21913, 3500.0,6.13510,
            4000.0,6.07040, 4500.0,6.01822, 5000.0,5.97432, 6000.0,5.90181,
            7000.0,5.84156, 8000.0,5.78889, 9000.0,5.74211,10000.0,5.70050,
           12000.0,5.62944,15000.0,5.54300,20000.0,5.43785,25000.0,5.35842,
           30000.0,5.29298,35000.0,5.23538,40000.0,5.18263,50000.0,5.08615,
            60000.0,4.99762,80000.0,4.83617,100000.0,4.68988,
                                      -1.0,0.0 ]
    c1=0.20
    c2=0.08
    c3=0.0065
    d=0.1
    i = 0
    while dist > d and d>0:
        d=ptable[i]
        i+=2
    if (d<1):
        return 0
    i+= -2
    d3=ptable[i]
    t3= d3/ptable[i+1]
    if d3 >= 50.0:
        d1=ptable[i-2]
        t1=d1/ptable[i-1]
    else:
        d1 = 0
        t1 = 0
    t = t1 + (t3-t1)*(dist-d1)/(d3-d1)
    v = dist/t
    t950 = t +c1+c2*v +c3*frac(dist)*v*v
    k = 0.0654 - 0.00258*v
    a = 85/k
    b = 1-950/a
    p = a*(t950/tsec - b)
    return p
```

```python
# Using meet as counter, fill in all relevant info from each competition
divcount = 1
performance_list = []
def performance_check(divcount):
    meet = selall.xpath('//*[@id="meet-results"]/div['+str(divcount)+']//th/a/text()').get()
    return meet

while performance_check(divcount):
    event = [i.strip() for i in selall.xpath\
                   ('//*[@id="meet-results"]/div['+str(divcount)+']/table//td[1]/text()').getall()]
    date = [i.strip() for i in selall.xpath\
            ('//*[@id="meet-results"]/div['+str(divcount)+']/table/thead/tr/th/span/text()').getall()]
    performance = selall.xpath('//*[@id="meet-results"]/div['+str(divcount)+']/table//tr/td//a/text()').getall()
    place = [i.replace(u'\xa0\n', u' ') for i in (i.strip() for i in selall.xpath\
             ('//*[@id="meet-results"]/div['+str(divcount)+']/table//tr/td[3]/text()').getall())]       
    dictionary = {'Location': performance_check(divcount), 'Event': event, 'Date': date,
                  'Performance': performance, 'Place': place}
    performance_list.append(dictionary)
    divcount += 1

# Match rows across
df = pd.DataFrame(performance_list)
date = df['Date'].apply(pd.Series).rename(columns = {0: 'Date'})
events = df['Event'].apply(pd.Series).rename(columns = lambda x : 'event')
performance = df['Performance'].apply(pd.Series).rename(columns = lambda x: 'performance')
place = df['Place'].apply(pd.Series).rename(columns = lambda x: 'place')
df = pd.concat([df['Location'], date, events, performance[:], place[:]], axis = 1)
event_df = df[['Location', 'Date', 'event']].set_index(['Location', 'Date']).stack()
event_df = event_df.reset_index().drop(columns = 'level_2').rename(columns={0: 'Event'})
perf_df = df[['Location', 'Date', 'performance']].set_index(['Location', 'Date']).stack()
perf_df = perf_df.reset_index().drop(columns = 'level_2').rename(columns={0: 'Performance'})
place_df = df[['Location', 'Date', 'place']].set_index(['Location', 'Date']).stack()
place_df = place_df.reset_index().drop(columns = 'level_2').rename(columns={0: 'Place'})

df = pd.concat([event_df, perf_df['Performance'], place_df['Place']], axis=1)

df['Name'], df['Team'] = \
[[i.strip().title().replace(u'\n\n', u' ') for i in selall.xpath\
('/html/body/form/div/div/div/div[1]/a[1]/h3/text()').getall()] * len(df),\
[i.strip().title() for i in selall.xpath('/html/body/form/div/div/div/div[1]/a[3]/h3/text()').getall()] *len(df)]

# Convert time to seconds
from datetime import datetime, timedelta
import re

date1 = pd.to_datetime(df["Performance"].map(lambda x: re.sub('[^.:0-9]','', x)), format="%M:%S.%f", errors = 'coerce')
second1 = date1.dt.minute * 60 + date1.dt.second + date1.dt.microsecond / 1000000
date2 = pd.to_datetime(df["Performance"].map(lambda x: re.sub('[^.:0-9]','', x)), format="%S.%f", errors = 'coerce')
second2 = date2.dt.minute * 60 + date2.dt.second + date2.dt.microsecond / 1000000
df["Performance_sec"] = second1.combine_first(second2)
df['Performance_sec'] = df.Performance_sec.astype(float, errors = 'ignore')

# Convert off distances to meters
df['Event_dist'] = pd.to_numeric(df['Event'].str.replace(',',''), errors = 'coerce')
df.loc[df.Event == 'Mile', 'Event_dist'] = 1609.34

# Plug into Purdy Points function
df['Purdy Points'] = np.vectorize(purdy)(df['Event_dist'], df['Performance_sec'])
df['Max Purdy Points'] = df['Purdy Points'].max(axis = 0)

# Get cohort based on first spring result
df['Cohort'] = df.loc[df.Date.str.contains('Jan|Feb|Mar|Apr|May', regex=True), 'Date'].values[-1][-4:]
df
```
